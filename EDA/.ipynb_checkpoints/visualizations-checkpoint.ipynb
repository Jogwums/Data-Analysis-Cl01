```python
#set up
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

print("Setup completed")
```

```python
x = np.arange(-10,11)
```

```python
#Mathematical Plotting library based off MATLAB...
#Matplotlibs default pyplot API, 
```

```python
plt.figure(figsize=(12,6))
plt.title("fancy plot")
plt.plot(x,x**2) # first plot 
plt.plot(x,-1*(x ** 2))
```

 Problems with global API 

```python
plt.figure(figsize=(12,6))
plt.title("my plot")

#the global api works depending on the order you write your code 
# you cannot specify which plot shows what 
plt.subplot(1,2,1) #rows, columns, panel selected 
# My figure will have one row, two columns, and I am going to start drawing in the first (left) plot.
plt.plot(x, x**2)
plt.plot([0,0,0], [-10,0,100])
plt.legend(['X^2', 'Vetical line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1,2,2) #without specifying this the plot wont be split
plt.plot(x, -1 *(x**2))
plt.plot([-10,0,10], [-50,-50,-50])
plt.legend(['-X^2', 'Horizontal line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

```python
# in the above example the plot flows top down
# it may be difficult to keep track of what is happening if you have say 100 lines of code 
# generating your plot 
```

Object Oriented API

```python
# figure and axes seperated
fig, axes = plt.subplots(figsize=(12,6)) #one axes 
```

```python
axes.plot(x, (x ** 2), color='red', linewidth=3, marker='o', markersize=8, label='X^2')

fig #shows the figure
```

```python
axes.plot(x, (x ** 2), color='red', linewidth=3, marker='o', markersize=8, label='X^2')

axes.plot(x, -1 * (x**2), 'b--', label='-X^2')
axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("my nice plot")

axes.legend()

fig
```
