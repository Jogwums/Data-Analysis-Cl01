```python
import pandas as pd
```

```python
url = './datasets/lrdata.csv'

df = pd.read_csv(url, delimiter=" ")

df
```

cross-tabulation (or crosstab for short) is a special case of a pivot table that computes group frequencies.

```python
pd.crosstab(df.Nationality, df.Handedness, margins=True)
```

The first two arguments to crosstab can each either be an array or Series or a list of
arrays.

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
pd.crosstab([tips.time, tips.day], tips.smoker, margins=True)
```
