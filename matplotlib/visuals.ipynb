# Visualization using Matplotlib 

```python
# matplotlib has two APIs 
# the global API and the Object oriented API
```

```python
# setup 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline

print("setup ready")
```

# Line charts

```python
# bring in the cars dataset 
url = "datasets/cars.csv"
cars = pd.read_csv(url)
```

```python
cars.head()
```

```python
# create a line chart 
# sALES VS YEARS 
```

```python
years = [2010, 2012, 2014, 2016, 2018, 2020]
Sales = [120, 125, 138, 146, 155, 164] # MILLIONS
```

```python
plt.figure(figsize=(10,5))
plt.plot(years, Sales, color="red", marker="o", linewidth=2, linestyle="dashed")
plt.title("Company X Sales")
plt.ylabel("millions of naira")
```

### Object oriented API

```python
fig, ax = plt.subplots(figsize=(12,6))
# change the background
ax.set_facecolor((0.4,0.4,0.5)) #rgb
plt.plot(years, Sales, color="white", marker="o", linewidth=2, linestyle="dashed")
plt.title("Company X Sales")
plt.ylabel("millions of naira")
```

# Histogram

```python
# used to show frequency distributions
# e.g height of 250 students
```

```python
x = np.random.normal(170, 10, 250)
# randomly generate an array with 250 values 
# values hould tend towards 170 and have a standard dev of 10 
```

```python
# print(x)
# draw a histogram 
fig, ax = plt.subplots(figsize=(12,6))
# change the background
ax.set_facecolor((0.2,0.6,0.2)) #rgb
plt.hist(x, color="white")
```

```python
plt.hist(y, 100) #bins 100
```

```python
cars.columns
```

```python
x = cars.Car
y = cars.CO2
z = cars.Weight 
```

```python
# global API 
plt.figure(figsize=(12,6))
plt.scatter(x,y)
plt.title("Cars vs CO2 Emmission")
```

```python
# weight vs co2
```

# Pie Chart

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23, 17, 35, 29, 12]
```

```python
plt.figure(figsize=(8,8))
plt.title("Class Distribution")
plt.pie(students, labels=langs, autopct='%1.2f%%')
plt.show() # show chart
```

```python
from matplotlib import cm
cs = cm.Set1(np.arange(5)/5)
```

```python
plt.pie(students, labels=langs, colors=cs, autopct='%1.2f%%')
plt.show()
```

# violin plot

```python
np.random.seed(10)

grp_1 = np.random.normal(100, 10, 200)
grp_2 = np.random.normal(80, 30, 200)
grp_3 = np.random.normal(90, 20, 200)
grp_4 = np.random.normal(70, 25, 200)
```

```python
data = [grp_1, grp_2, grp_3, grp_4]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.violinplot(data)
plt.show()
```

# Box plot

```python
plt.boxplot(data)
plt.show()
```
