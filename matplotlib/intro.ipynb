```python
from matplotlib import pyplot as plt 
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python

#create a line chart 
#years on the x-axis, gdp on y-axis
```

# Line Charts

```python
plt.plot(years, gdp)
plt.show()
```

```python
#improve the chart 
```

```python
plt.plot(years, gdp, color='green') #change color 
plt.show()
```

```python
plt.plot(years, gdp, color='red', marker='x') #marker (o,X,p) adds the string to the data points
plt.show()
```

```python
plt.plot(years, gdp, color='red', marker='o', linestyle='dotted') #linestyle (solid, dashed, dotted)
plt.show()
```

```python
#add a title and x and y axis labels 
plt.plot(years, gdp, color='red', marker='o', linestyle='dashed') 
plt.title('Nominal GDP')
plt.ylabel('Billions of dollars $')
plt.xlabel('Years') #Not needed as it's self descriptive 
plt.show()
```

```python
fig, ax = plt.subplots()
ax.set_facecolor((0.1, 0.2, 0.5)) #color specification in matplotlib
plt.title('Nominal GDP') # add title 
plt.ylabel("billions of dollars $")
# plt.xlabel('Years')
ax.plot(years, gdp, color=(1.0, 1.0, 1.0), marker='X', linestyle='dashed')
```

# Bar Charts 

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]
```

```python
# bars are by default width 0.8, so we'll add 0.1 to the left coordinates
# so that each bar is centered
```

```python
plt.bar(movies, num_oscars)
```

```python
xs = [i + 0.1 for i, _ in enumerate(movies)]
```

```python
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.bar(xs, num_oscars)
```

```python
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
```

```python
# plot bars with left x-coordinates [xs], heights [num_oscars]
plt.bar(xs, num_oscars)
plt.ylabel("# of Academy Awards")
plt.title("My Favorite Movies")
# label x-axis with movie names at bar centers
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)
plt.show()
```
