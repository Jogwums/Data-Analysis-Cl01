## Filtering 

```python
df2.loc[2012:2014, 'Make': '(kW)']
#will return rows 2012 to 2014 and cols Make to kW
```

```python
#return specific columns 
df2.loc[2012:2014, ['Make', 'Model', 'TYPE']].head()
```

```python
df2.loc[[2012,2014,2016], ['Make', 'Model', 'TYPE', '(kW)']].tail()
```

```python
df2.loc[df2['Make']=='FORD']
```

```python
df2[df2['Make']=='FORD']
```

```python
df2.loc[df2['Make']=='FORD', 'Make':'TYPE']
```

```python
df2[df2['Make']=='TESLA']
```

```python
#drilling down
df2[(df2['Make']=='TESLA') &(df2['Model']=='MODEL S 70D')]
```

```python
#return specific columns
df2.loc[(df2['Make']=='TESLA') &(df2['Model']=='MODEL S 70D'), ['Make','Model', 'Size', '(kW)']]
```

## lambda functions 

```python
df.loc[df['Make'].apply(lambda x: len(x)==5)]
```

```python
#the above code returns any make with a 5 word length only
```

```python
#filtering and returning specific columns 
df.loc[df['Make'].apply(lambda x: len(x)==5), 'Make':'TYPE']
```
