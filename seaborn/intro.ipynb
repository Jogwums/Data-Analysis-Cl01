```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

print('Setup Done')
```

```python
# import data 
tips = sns.load_dataset('tips')
tips.head()
```

```python
# plot some columns of the data 
```

```python
# use categorical plots 
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
# remove jitter
sns.catplot(x='day', y='total_bill', jitter=False, data=tips)
```

```python
# change plot to swarm plot 
sns.catplot(x='day', y='total_bill', data=tips, kind="swarm")
```

```python
# add another dimension to a categorical plot using the Hue semantic 
sns.catplot(x='day', y='total_bill', data=tips, kind="swarm", hue='sex')
```

```python
# ploting the size column 
sns.catplot(x='size', y='total_bill', data=tips)
```

```python
#Diamonds Data 
diamonds = sns.load_dataset('diamonds')
diamonds
```

```python
# large dataset with info on different diamonds
# carat, cut, color, clarity etc
# use wiki to explain these features 
```

```python
sns.lmplot(x='price',y='carat', data=diamonds)
```

```python
# there is a line under the scatter plot 
# in order to check these clearly we can 
# shuffle the dataset, then get the 1st 100 rows
```

```python
df = sns.load_dataset('diamonds').sample(frac=1).head(100)
df
```

```python
# draw a regression plot of this smaller sample
sns.lmplot(x='price',y='carat', data=df)
```

```python
# easier to see the line of regression now 
```

```python
# modify thee plot 
```

```python
sns.lmplot(x='price',y='carat', data=df, scatter_kws={'marker':'x', 'color':'purple'}, line_kws={'color':'red','linewidth':2})
```

## higher order trend lines 

```python
sns.lmplot(x='price',y='carat',order=2, data=df, scatter_kws={'marker':'x', 'color':'purple'}, line_kws={'color':'red','linewidth':2})
```

```python
sns.lmplot(x='price',y='carat',order=4, data=df, scatter_kws={'marker':'x', 'color':'purple'}, line_kws={'color':'red','linewidth':2})
```

```python
# Plotting without displaying a trendline
# set fit_reg = False 
sns.lmplot(x='price',y='carat', data=df, fit_reg=False, scatter_kws={'marker':'x', 'color':'purple'},)
```

```python
# more arguments
# hue
plt.figure(figsize=(14,8))
sns.lmplot(x='price',y='carat', data=df, fit_reg=False, hue='cut',)
```

```python
# specifying multiple markers
sns.lmplot(x='price',y='carat', hue='cut', markers=['o','x','^','*','v'], data=df,)
```

```python
sns.lmplot(x='price',y='carat', data=df,hue='cut')
```

```python
# local regression
"""
Local regression or local polynomial regression,[1] also known as moving regression,[2] is a generalization of moving average and polynomial regression.[3] Its most common methods, initially developed for scatterplot smoothing, are LOESS (locally estimated scatterplot smoothing) and LOWESS (locally weighted scatterplot smoothing), both pronounced /los/. They are two strongly related non-parametric regression methods that combine multiple regression models in a k-nearest-neighbor-based meta-model. Outside econometrics, LOESS is known and commonly referred to as SavitzkyGolay filter [4][5] (proposed 15 years before LOESS).
"""
```

```python
sns.set_theme(style='darkgrid')
sns.lmplot(x='price',y='carat',data=df, lowess=True)
```

```python
#regplot
sns.regplot(x='price',y='carat',data=df)
```
